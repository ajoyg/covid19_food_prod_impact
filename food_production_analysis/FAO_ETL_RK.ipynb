```python
#import dependencies
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# establish a connection to the data
FAO_data_path = Path("Resources/FAO_food_price_index.csv")

#read data and store as a pandas data frame
FAO_data = pd.read_csv(FAO_data_path)

#drop empty columns
FAO_data = FAO_data.dropna(axis=1, how='all')
FAO_data = FAO_data.drop(["Unnamed: 36", "Unnamed: 65"], axis=1)

FAO_data
```

```python
#Look for missing data
FAO_data.count()
```

```python
#Drop rows with missing data
FAO_data = FAO_data.dropna(how='any')

FAO_data
```

```python
#Save cleaned data to "staging" folder

FAO_data.to_csv("staging/FAO_data_cleaned", index=False)
```
