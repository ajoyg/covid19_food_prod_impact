## Staging the CPI data from 2018 to 2022 for food items

```python
import csv
import pandas as pd
```

```python
#Read the CSV file and prepare to add it to a dataframe
csvpath="Resources/BLS_CPI_FOOD.csv"
#open the BLS_CPI_FOOD.csv file
with open(csvpath, encoding='utf8') as CPI_FOOD_DATA:
    food_price_reader = csv.reader(CPI_FOOD_DATA, delimiter=',')
    series_id = None
    series_title = None
    area = None
    item = None
    price_data = []
    for row in food_price_reader:
        # If the row starts with "Series Id:", "Series Title:", "Area:", or "Item:",
        # extract the corresponding information and continue to the next row
        if row:
            if row[0].startswith('Series Id:'):
                series_id = row[0].split(':')[1].strip()
            elif row[0].startswith('Series Title:'):
                row[0] = row[0].split(':')[1].strip()
                series_title = ''.join(row)
            elif row[0].startswith('Area:'):
                area = row[0].split(':')[1].strip()
            elif row[0].startswith('Item:'):
                row[0] = row[0].split(':')[1].strip()
                item=''.join(row)
            elif row[0].isdigit():
                # If the first column is a digit, assume it's a year and extract the prices
                prices = [p if p != 'No Data Available for selected year(s)' else None for p in row[0:]]
                prices.insert(0,item)
                price_data.append(prices)
#Add the dataset to a datafrane
item_price_data_df = pd.DataFrame(price_data, columns=['Item','Year','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
item_price_data_df.set_index("Item", inplace=True)
    
```

```python
#Save the item price CPI data into a csv in the staging folder
item_price_data_df.to_csv("staging/food_cpi.csv")
```
