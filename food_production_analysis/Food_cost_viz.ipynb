```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
item_price_data_df = pd.read_csv("staging/food_cpi.csv")
item_price_data_df.head(10)
```

```python
item_price_data_df['Average Price'] = item_price_data_df[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']].mean(axis=1)
item_price_data_df
```

```python
#Extract columns to plot
item_avg_yearly_price = item_price_data_df[['Item','Year','Average Price']]
item_avg_yearly_price
```

```python
item_avg_graph = item_avg_yearly_price.pivot(index='Year', columns='Item', values='Average Price')
ax = item_avg_graph.plot(figsize= (20,10))
ax.legend(bbox_to_anchor=(1.0,1.0))
plt.xlabel("Years")
plt.ylabel("Food Item Cost ($)")
plt.xticks(np.arange(item_avg_yearly_price['Year'].min(), item_avg_yearly_price['Year'].max()+1, 1.0))
ax.plot()
```

```python
fao_price_index_df = pd.read_csv("staging/FAO_data_cleaned")
fao_price_index_df
```

```python
#Filter rows between 2018 to 2022
fao_price_index_df['Year'] = fao_price_index_df['Year'].astype(str)
fao_price_index_df['Year']=pd.to_numeric(fao_price_index_df['Date'].str.slice(0,4))
fao_price_idx_5_years= fao_price_index_df.loc[(fao_price_index_df['Year'] > 2017) & (fao_price_index_df['Year'] < 2023),:]
```

```python
fao_final_price_index=fao_price_idx_5_years.groupby(['Year']).agg({'Food Price Index':"mean",'Meat':"mean",'Dairy':"mean",'Cereals':"mean",'Oils':"mean",'Sugar':'mean'})
fao_final_price_index
```

```python
years = fao_final_price_index.index
food_price_index,= plt.plot(years, fao_final_price_index['Food Price Index'], color="green", label="Food Price Index")
meat,= plt.plot(years, fao_final_price_index['Meat'], color="red", label="Meat")
dairy,= plt.plot(years, fao_final_price_index['Dairy'], color="blue", label="Dairy")
cereals,= plt.plot(years, fao_final_price_index['Cereals'], color="orange", label="Cereals")
oils,= plt.plot(years, fao_final_price_index['Oils'], color="yellow", label="Oils")
sugar,= plt.plot(years, fao_final_price_index['Sugar'], color="pink", label="Sugar")
plt.xlabel("Years")
plt.xticks(np.arange(min(years), max(years)+1, 1.0))
plt.legend(loc="best")
plt.show()

```
