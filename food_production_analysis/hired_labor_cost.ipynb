```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Get the farm labor cost data
hired_labor_cost=pd.read_csv("Resources/USDA_US_Labor_costs_2016_2021.csv")
hired_labor_cost
```

```python
#Build the charts for the selected grain commodities exclude rice
years=hired_labor_cost['Year']
corn_labor_price,= plt.plot(years, hired_labor_cost['Corn'], color="yellow", label="Corn labor price/acre")
barley_labor_price,= plt.plot(years, hired_labor_cost['Barley'], color="brown", label="Barley labor price/acre")
#rice_labor_price,= plt.plot(years, hired_labor_cost['Rice'], color="red", label="Rice labor price/acre")
sorghum_labor_price,= plt.plot(years, hired_labor_cost['Sorghum'], color="orange", label="Sorghum labor price/acre")
oats_labor_price,= plt.plot(years, hired_labor_cost['Oats'], color="blue", label="Oats labor price/acre")
soybeans_labor_price,= plt.plot(years, hired_labor_cost['Soybeans'], color="pink", label="Soybeans labor price/acre")
wheat_labor_price,= plt.plot(years, hired_labor_cost['Wheat'], color="green", label="Wheat labor price/acre")
plt.xlabel("Years")
plt.ylabel("Price per acre")
plt.legend(loc="best")
plt.title("Labor Price - Grains")
plt.savefig("output/grain_labor_price.png")
plt.show()
```

```python
#Build the chart for milk labor price

milk_labor_price,= plt.plot(years, hired_labor_cost['Milk'], color="blue", label="Milk labor price $/100 weight")
plt.xlabel("Years")
plt.legend(loc="best")
plt.ylabel("$ per Hunderd Weight")
plt.title("Milk Labor Price")
plt.savefig("output/milk_labor_price.png")
plt.show()
```

```python
#Build the chart for beef labor price
beef_labor_price,= plt.plot(years, hired_labor_cost['Cow'], color="blue", label="Beef labor price $/cow")
plt.xlabel("Years")
plt.legend(loc="best")
plt.title("Beef Labor Price")
plt.savefig("output/beef_labor_price.png")
plt.show()
```

```python
#Build the chart for pork labor price
pork_labor_price,= plt.plot(years, hired_labor_cost['Hog'], color="blue", label="Pork labor price $/100 weight gain")
plt.xlabel("Years")
plt.legend(loc="best")
plt.title("Pork Labor Price")
plt.savefig("output/pork_labor_price.png")
plt.show()
```
